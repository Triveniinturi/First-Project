```python
import pandas as pd
import numpy as np

# Path to your file
csv_path = "/content/drive/MyDrive/Colab Notebooks/YoloProject/Food-Calorie-Recommender-app/runs/detect/train/results.csv" #epoch 150 results.csv

# Read CSV (allow bad lines to be skipped)
df = pd.read_csv(csv_path, on_bad_lines='skip')
df.columns = df.columns.str.strip()

# Remove any rows that are headers or text
df = df[df["epoch"].astype(str).str.isnumeric()]

# Convert all columns (except maybe learning rates) to numeric safely
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Drop rows with any NaN epoch (just in case)
df = df.dropna(subset=["epoch"])
df["epoch"] = df["epoch"].astype(int)

print("✅ Cleaned data shape:", df.shape)
print("Epoch range:", df["epoch"].min(), "→", df["epoch"].max())

# Now compute losses safely
train_loss = df["train/box_loss"] + df["train/cls_loss"] + df["train/dfl_loss"]
val_loss = df["val/box_loss"] + df["val/cls_loss"] + df["val/dfl_loss"]

# Plot again
import matplotlib.pyplot as plt

plt.figure(figsize=(9,5))
plt.plot(df["epoch"], train_loss, label="Training Loss", color="blue", linewidth=2)
plt.plot(df["epoch"], val_loss, label="Validation Loss", color="red", linewidth=2)
plt.title("Training vs Validation Loss — YOLOv8 (Cleaned Data)")
plt.xlabel("Epochs")
plt.ylabel("Loss Value")
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(9,5))
plt.plot(df["epoch"], df["metrics/mAP50(B)"], label="mAP@50", color="green")
plt.plot(df["epoch"], df["metrics/mAP50-95(B)"], label="mAP@50-95", color="orange")
plt.title("Validation mAP across Epochs")
plt.xlabel("Epochs")
plt.ylabel("mAP")
plt.legend()
plt.grid(True)
plt.show()


# Compute total train & validation loss
train_loss = df["train/box_loss"] + df["train/cls_loss"] + df["train/dfl_loss"]
val_loss = df["val/box_loss"] + df["val/cls_loss"] + df["val/dfl_loss"]

# Compute moving averages (for smoother trend)
train_smooth = train_loss.rolling(window=5, min_periods=1).mean()
val_smooth = val_loss.rolling(window=5, min_periods=1).mean()

# Find epoch where validation loss stops improving for a while
best_epoch = np.argmin(val_smooth.values) + 1
overfit_epoch = None

for i in range(best_epoch, len(df)):
    # if validation loss increases for 5 consecutive epochs while training loss decreases
    if all(val_smooth[i-5:i].diff().dropna() > 0) and all(train_smooth[i-5:i].diff().dropna() < 0):
        overfit_epoch = df["epoch"].iloc[i]
        break

# Plot
plt.figure(figsize=(10,6))
plt.plot(df["epoch"], train_smooth, label="Training Loss (smoothed)", color="blue", linewidth=2)
plt.plot(df["epoch"], val_smooth, label="Validation Loss (smoothed)", color="red", linewidth=2)
plt.axvline(best_epoch, color='green', linestyle='--', label=f'Best Val Epoch ({best_epoch})')

if overfit_epoch:
    plt.axvline(overfit_epoch, color='orange', linestyle='--', label=f'Overfitting Starts (~Epoch {overfit_epoch})')

plt.title("YOLOv8 — Training vs Validation Loss (Overfitting Detection)")
plt.xlabel("Epochs")
plt.ylabel("Loss Value")
plt.legend()
plt.grid(True)
plt.show()

print(f"📉 Best validation loss at epoch {best_epoch}")
if overfit_epoch:
    print(f"⚠️ Overfitting likely starts around epoch {overfit_epoch}")
else:
    print("✅ No strong overfitting pattern detected — training looks balanced.")
```
